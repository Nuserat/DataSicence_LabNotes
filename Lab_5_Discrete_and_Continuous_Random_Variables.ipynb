# Part 1: Exploring Discrete Random Variables

## Simulating Dice Rolls
Objective: Simulate rolling a six-sided die 1000 times and plot the frequency of each outcome.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate rolling a six-sided die 1000 times
rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)

rolls
```

```python
# Plot the frequency of each outcome
plt.hist(rolls, bins=np.arange(1, 8)-0.5, density=True, rwidth=0.8)
plt.title('Frequency of Dice Rolls')
plt.xlabel('Die Outcome')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))
plt.show()
```

## Binomial Distribution Analysis
Objective: Simulate a binomial distribution for flipping a coin 10 times in 1000 trials, plot it, and calculate statistical properties.

```python
from scipy.stats import binom

# Parameters for the binomial distribution
n, p = 10, 0.5  # number of trials, probability of success (head) in each trial

# Simulate a binomial distribution
data_binom = binom.rvs(n, p, size=1000)


```

```python
# Plot the distribution
plt.hist(data_binom, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.75)
plt.title('Binomial Distribution - Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()

```

```python
# Calculate mean and standard deviation
mean, var = binom.stats(n, p)
std = np.sqrt(var)
print(f'Mean: {mean}, Standard Deviation: {std}')
```

# Part 2: Delving into Continuous Random Variables

## Visualizing Normal Distributions
Objective: Generate and visualize three normal distributions with different means and standard deviations.

```python
# Generate normal distributions
mean_std_values = [(0, 1), (2, 1.5), (-2, 0.5)]
x = np.linspace(-6, 6, 1000)
```

```python
# Plot distributions
for mean, std in mean_std_values:
    plt.plot(x, np.exp(-(x - mean)**2 / (2 * std**2)) / (np.sqrt(2 * np.pi) * std), label=f'mean {mean}, std {std}')

plt.title('Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## Analyzing Exponential Distributions
Objective: Simulate an exponential distribution for the time between bus arrivals and analyze the waiting time probability.

```python
from scipy.stats import expon

# Simulate an exponential distribution
mean_waiting_time = 10  # Mean waiting time of 10 minutes
data_expon = expon.rvs(scale=mean_waiting_time, size=1000)
```

```python
# Plot the distribution
plt.hist(data_expon, bins=30, density=True, alpha=0.6)
plt.title('Exponential Distribution - Bus Arrival Times')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.show()

```

```python
# Calculate the probability of waiting longer than 15 minutes
prob_over_15 = 1 - expon.cdf(15, scale=mean_waiting_time)
print(f'Probability of waiting longer than 15 minutes: {prob_over_15}')
```

## Fit Distributions & Visualize Fits
We will focus on the BMI (Before COVID) and BMI (During COVID) as our primary continuous variables for distribution fitting. Let's visualize these distributions first:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the distribution of BMI Before and During COVID
plt.figure(figsize=(12, 6))
sns.kdeplot(df['BMI (Before COVID)'], label='BMI Before COVID', fill=True)
sns.kdeplot(df['BMI (During COVID)'], label='BMI During COVID', fill=True)
plt.title('Distribution of BMI Before and During COVID')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Calculate Statistical Measures & Interpret

```python
# Calculate and print statistical measures for BMI Before COVID
bmi_before_mean = df['BMI (Before COVID)'].mean()
bmi_before_median = df['BMI (Before COVID)'].median()
bmi_before_std = df['BMI (Before COVID)'].std()

print(f'BMI Before COVID - Mean: {bmi_before_mean}, Median: {bmi_before_median}, Standard Deviation: {bmi_before_std}')

# Repeat for BMI During COVID
bmi_during_mean = df['BMI (During COVID)'].mean()
bmi_during_median = df['BMI (During COVID)'].median()
bmi_during_std = df['BMI (During COVID)'].std()

print(f'BMI During COVID - Mean: {bmi_during_mean}, Median: {bmi_during_median}, Standard Deviation: {bmi_during_std}')
```
